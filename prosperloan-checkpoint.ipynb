## Univariate Exploration


```python

def hist_chart(pld_df,xVar):
    # set figsize
    plt.figure(figsize = [14,8])
    bins = 50
    #set title
    plt.title(f'''Distribution{xVar}''' .title(),weight='bold')
    
    #set plot
    plt.hist(data=pld_df,x = xVar,bins = bins)
    
    # set labels for X and Y
    plt.xlabel(xVar.title(),weight='bold')
    plt.ylabel('Count'.title(),weight = 'bold')
    fontname="Arial"
    
```

```python
hist_chart(pld_df,'BorrowerRate')
```

```python
hist_chart(pld_df,'BorrowerAPR')
```

> The borrower rate or interest has a near normal distibution, with a high spike between 0.1 and 0.2. There's also another high spike in the 0.3 region.

> The borrowerAPR which is also the cost of borrowing has also a near normal distribution with the highest peak being at 35%

```python
def bar_countplot(pld_df,xVar):
    """
    A function that plots count plots
    """
    plt.figure(figsize = [14,8])
    color_base = sb.color_palette()[9]
    plt.title(f'''Distibution{xVar}''' .title(), weight='bold')
    plt.xlabel(xVar.title(),weight='bold')
    plt.ylabel('Count'.title(),weight='bold')
    fontname='Arial'
    sb.countplot(data = pld_df,x = xVar, color = color_base )
    
```

```python
# plot for ProsperScore
bar_countplot(pld_df,'ProsperScore')
```

> The Prosper score is a tool built using historical data it's a custom risk score that categorises borrower risk. It ranges from 1-11, 11 being the best or lowest score. For risks between 1-3 considered to be high risks recorded increamental risks. Between 4-8 the borrowing risk increases but its average, it decreases from 9-11.

The next univariate variables are to be converted to categorical ordinal variables

```python
# plot for loan category
bar_countplot(pld_df,'ListingCategory (numeric)')
```

```python
# convert income range and prosper rating alpha into ordered categorical type
ordinal_cat = {'ProsperRating (Alpha)':['N/A','HR','E','D','C','B','A','AA'],
               'IncomeRange':['$0','$1-24,999','$25,000-49,999','$50,000-74,999','$75,000-99,999','$100,000+'],'LoanStatus':
              ['Current','Completed','FinalPaymentinProgress','Past Due (1-15 days)','Past Due (31-60 days)',
              'Past Due (61-90 days)','Past Due (91-120 days)','Past Due (>120 days)','Chargedoff','Defaulted']}

for cat in ordinal_cat:
    ordered_cat = pd.api.types.CategoricalDtype(ordered = True, categories = ordinal_cat[cat])
    
    pld_df[cat] = pld_df[cat].astype(ordered_cat)
    
```

```python
# plot for loan status
plt.figure(figsize = [14,8])
color_base = sb.color_palette()[9]
sb.countplot(data = pld_df,y= 'LoanStatus',color = color_base);
plt.title('Loan status Distribution');
```

> The current loan status which means, a loan that is fully paid to date according to a contract is the highest. It's followed by completed which means, a loan which means a waived by the lender and agreed between parties. Charged off is a loan status is a loan written as a loss and account is closed for future charges. Default loan is a status where borrower fails to pay back debt according to initial aggrement. The past due status is the lowest status of the category, it's simply a payment that has not been made by its cutoff time at the end of its due date. 

> Majority of borrowers are in the completed and current loan status which is a good thing, for the other three statues past due, defaulted and charged off represent a fairly small number of borrowers but cannot be ignored if they are on the upward trend.

```python
def any_boxplot(pld_df,xVar,yVar):
    plt.figure(figsize = [14,8])
    plt.title(f'Relationship between {yVar} and {xVar} ',weight='bold')
    plt.xlabel(xVar.title())
    plt.ylabel(yVar.title())
    sb.boxplot(data = pld_df, x = xVar, y = yVar)
    
```

```python
any_boxplot(pld_df,'ListingCategory (numeric)','BorrowerRate')
```

> Listing Category: The category of the listing that the borrower selected when posting their listing: 0 - Not Available, 1 - Debt Consolidation, 2 - Home Improvement, 3 - Business, 4 - Personal Loan, 5 - Student Use, 6 - Auto, 7- Other, 8 - Baby&Adoption, 9 - Boat, 10 - Cosmetic Procedure, 11 - Engagement Ring, 12 - Green Loans, 13 - Household Expenses, 14 - Large Purchases, 15 - Medical/Dental, 16 - Motorcycle, 17 - RV, 18 - Taxes, 19 - Vacation, 20 - Wedding Loans
